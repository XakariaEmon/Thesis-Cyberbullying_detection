```python
class CyberbullyingDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item
```

```python

from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score
from tqdm import tqdm
```

```python
def evaluate_model(model, test_dataset, batch_size=16):
    model.eval()  # Set model to evaluation mode
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    all_preds = []
    all_labels = []
    
    with torch.no_grad():  # Disable gradient computation for inference
        for batch in tqdm(test_loader, desc="Evaluating"):
            # Move inputs to device (CPU/GPU)
            inputs = {key: val.to(device) for key, val in batch.items() if key != "labels"}
            labels = batch["labels"].to(device)
            
            outputs = model(**inputs)
            logits = outputs.logits  # Get raw scores
            preds = torch.argmax(logits, dim=1)  # Get class with highest probability
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Compute accuracy
    accuracy = accuracy_score(all_labels, all_preds)
    return accuracy
```

```python
# Define the ensemble prediction function
def ensemble_predict(bert_model, roberta_model, bert_dataset, roberta_dataset, albert_model, albert_dataset, gpt2_model, gpt2_dataset, xlnet_model, xlnet_dataset, batch_size=16):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    bert_model.to(device)
    roberta_model.to(device)
    albert_model.to(device)
    gpt2_model.to(device)
    xlnet_model.to(device)


    bert_loader = DataLoader(bert_dataset, batch_size=batch_size, shuffle=False)
    roberta_loader = DataLoader(roberta_dataset, batch_size=batch_size, shuffle=False)
    albert_loader = DataLoader(albert_dataset, batch_size=batch_size, shuffle=False)
    gpt2_loader = DataLoader(gpt2_dataset, batch_size=batch_size, shuffle=False)
    xlnet_loader = DataLoader(xlnet_dataset, batch_size=batch_size, shuffle=False)

    all_preds_hard = []
    all_preds_soft = []

    for (bert_batch, roberta_batch, albert_batch, gpt2_batch, xlnet_batch) in zip(bert_loader, roberta_loader, albert_loader, gpt2_loader, xlnet_loader):
        bert_inputs = {key: val.to(device) for key, val in bert_batch.items() if key != "labels"}
        roberta_inputs = {key: val.to(device) for key, val in roberta_batch.items() if key != "labels"}
        albert_inputs = {key: val.to(device) for key, val in albert_batch.items() if key != "labels"}
        gpt2_inputs = {key: val.to(device) for key, val in gpt2_batch.items() if key != "labels"}
        xlnet_inputs = {key: val.to(device) for key, val in xlnet_batch.items() if key != "labels"}
        

        with torch.no_grad():
            bert_outputs = bert_model(**bert_inputs).logits
            roberta_outputs = roberta_model(**roberta_inputs).logits
            albert_outputs = albert_model(**albert_inputs).logits
            gpt2_outputs = gpt2_model(**gpt2_inputs).logits
            xlnet_outputs = xlnet_model(**xlnet_inputs).logits
            
            bert_probs = F.softmax(bert_outputs, dim=1)
            roberta_probs = F.softmax(roberta_outputs, dim=1)
            albert_probs = F.softmax(albert_outputs, dim=1)
            gpt2_probs = F.softmax(gpt2_outputs, dim=1)
            xlnet_probs = F.softmax(xlnet_outputs, dim=1)
            
            avg_probs = (bert_probs + roberta_probs + albert_probs + gpt2_probs + xlnet_probs) / 5
            final_preds_soft = torch.argmax(avg_probs, dim=1).cpu().numpy()
            
            bert_preds = torch.argmax(bert_probs, dim=1).cpu().numpy()
            roberta_preds = torch.argmax(roberta_probs, dim=1).cpu().numpy()
            albert_preds = torch.argmax(albert_probs, dim=1).cpu().numpy()
            gpt2_preds = torch.argmax(gpt2_probs, dim=1).cpu().numpy()
            xlnet_preds = torch.argmax(xlnet_probs, dim=1).cpu().numpy()
            
            final_preds_hard = [Counter([b, r, a, g, x]).most_common(1)[0][0] for b, r, a, g, x in zip(bert_preds, roberta_preds, albert_preds, gpt2_preds, xlnet_preds)]


            all_preds_soft.extend(final_preds_soft)
            all_preds_hard.extend(final_preds_hard)

    return all_preds_hard, all_preds_soft
```

```python
import torch
import torch.nn.functional as F
from collections import Counter
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def weighted_ensemble_predict(bert_model, roberta_model, bert_dataset, roberta_dataset, albert_model, albert_dataset, gpt2_model, gpt2_dataset, xlnet_model, xlnet_dataset, batch_size=16):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Move models to device
    bert_model.to(device)
    roberta_model.to(device)
    albert_model.to(device)
    gpt2_model.to(device)
    xlnet_model.to(device)

    # Define model weights based on accuracy (you can adjust these based on your actual results)
    weights = {
        "bert": 0.8673,   # BERT accuracy
        "roberta": 0.8662,  # RoBERTa accuracy
        "albert": 0.8535,  # ALBERT accuracy
        "gpt2": 0.8558,   # GPT-2 accuracy
        "xlnet": 0.8642    # XLNet accuracy
    }

    # Create DataLoader for each dataset
    bert_loader = DataLoader(bert_dataset, batch_size=batch_size, shuffle=False)
    roberta_loader = DataLoader(roberta_dataset, batch_size=batch_size, shuffle=False)
    albert_loader = DataLoader(albert_dataset, batch_size=batch_size, shuffle=False)
    gpt2_loader = DataLoader(gpt2_dataset, batch_size=batch_size, shuffle=False)
    xlnet_loader = DataLoader(xlnet_dataset, batch_size=batch_size, shuffle=False)

    all_preds_hard = []  # List to store final hard predictions
    all_preds_soft = []  # List to store final soft predictions

    # Loop through each batch
    for (bert_batch, roberta_batch, albert_batch, gpt2_batch, xlnet_batch) in zip(bert_loader, roberta_loader, albert_loader, gpt2_loader, xlnet_loader):
        
        # Move batches to device
        bert_inputs = {key: val.to(device) for key, val in bert_batch.items() if key != "labels"}
        roberta_inputs = {key: val.to(device) for key, val in roberta_batch.items() if key != "labels"}
        albert_inputs = {key: val.to(device) for key, val in albert_batch.items() if key != "labels"}
        gpt2_inputs = {key: val.to(device) for key, val in gpt2_batch.items() if key != "labels"}
        xlnet_inputs = {key: val.to(device) for key, val in xlnet_batch.items() if key != "labels"}

        # Disable gradient calculation for inference
        with torch.no_grad():
            # Get outputs (logits) for each model
            bert_outputs = bert_model(**bert_inputs).logits
            roberta_outputs = roberta_model(**roberta_inputs).logits
            albert_outputs = albert_model(**albert_inputs).logits
            gpt2_outputs = gpt2_model(**gpt2_inputs).logits
            xlnet_outputs = xlnet_model(**xlnet_inputs).logits

            # Apply softmax to get probabilities
            bert_probs = F.softmax(bert_outputs, dim=1)
            roberta_probs = F.softmax(roberta_outputs, dim=1)
            albert_probs = F.softmax(albert_outputs, dim=1)
            gpt2_probs = F.softmax(gpt2_outputs, dim=1)
            xlnet_probs = F.softmax(xlnet_outputs, dim=1)

            # Calculate the weighted average probabilities for soft voting
            avg_probs = (
                weights["bert"] * bert_probs +
                weights["roberta"] * roberta_probs +
                weights["albert"] * albert_probs +
                weights["gpt2"] * gpt2_probs +
                weights["xlnet"] * xlnet_probs
            ) / sum(weights.values())  # Normalize by the total sum of weights
            final_preds_soft = torch.argmax(avg_probs, dim=1).cpu().numpy()

            # Individual predictions for hard voting (majority voting)
            bert_preds = torch.argmax(bert_probs, dim=1).cpu().numpy()
            roberta_preds = torch.argmax(roberta_probs, dim=1).cpu().numpy()
            albert_preds = torch.argmax(albert_probs, dim=1).cpu().numpy()
            gpt2_preds = torch.argmax(gpt2_probs, dim=1).cpu().numpy()
            xlnet_preds = torch.argmax(xlnet_probs, dim=1).cpu().numpy()

            # Hard voting based on weighted majority class
            final_preds_hard = [
                Counter([b, r, a, g, x]).most_common(1)[0][0]
                for b, r, a, g, x in zip(bert_preds, roberta_preds, albert_preds, gpt2_preds, xlnet_preds)
            ]

            # Store results
            all_preds_soft.extend(final_preds_soft)
            all_preds_hard.extend(final_preds_hard)

    return all_preds_hard, all_preds_soft
```

```python
# Assuming your models (BERT, RoBERTa, ALBERT, GPT-2, XLNet) and datasets (bert_test_dataset, etc.) are already defined and loaded

# Example test labels (replace with actual labels)
y_true = test_labels  # Replace with your actual test labels

# Call the weighted ensemble prediction function
hard_preds, soft_preds = weighted_ensemble_predict(
    bert_model, roberta_model, bert_test_dataset, roberta_test_dataset, 
    albert_model, albert_test_dataset, gpt2_model, gpt2_test_dataset, 
    xlnet_model, xlnet_test_dataset, batch_size=16
)

# Now evaluate the performance of the ensemble
accuracy = accuracy_score(y_true, hard_preds)
print(f"Ensemble Hard Voting Accuracy: {accuracy:.4f}")

# Precision, Recall, F1 Score (Weighted)
precision, recall, f1, _ = precision_recall_fscore_support(y_true, hard_preds, average='weighted')
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
import torch
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve, auc, matthews_corrcoef, cohen_kappa_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Assume these are already loaded:
# bert_model, roberta_model, albert_model, gpt2_model, xlnet_model
# bert_test_dataset, roberta_test_dataset, albert_test_dataset, gpt2_test_dataset, xlnet_test_dataset
# test_labels (PyTorch tensor or NumPy array)

device = torch.device('cpu')

# Define models and datasets
models = {
    'BERT': bert_model,
    'RoBERTa': roberta_model,
    'ALBERT': albert_model,
    'GPT2': gpt2_model,
    'XLNet': xlnet_model,
}

datasets = {
    'BERT': bert_test_dataset,
    'RoBERTa': roberta_test_dataset,
    'ALBERT': albert_test_dataset,
    'GPT2': gpt2_test_dataset,
    'XLNet': xlnet_test_dataset,
}

def get_predictions(model, dataset):
    loader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=False)
    model.to(device)
    model.eval()
    all_probs = []
    all_preds = []
    with torch.no_grad():
        for batch in loader:
            inputs = {k: v.to(device) for k, v in batch.items() if k != 'labels'}
            outputs = model(**inputs)
            logits = outputs.logits
            probs = torch.softmax(logits, dim=1)
            preds = torch.argmax(probs, dim=1)
            all_probs.append(probs.cpu().numpy())
            all_preds.append(preds.cpu().numpy())
    all_probs = np.concatenate(all_probs)
    all_preds = np.concatenate(all_preds)
    return all_probs, all_preds

# Get predictions and probabilities for each model
model_probs = []
model_preds = []

for name in models:
    probs, preds = get_predictions(models[name], datasets[name])
    model_probs.append(probs)
    model_preds.append(preds)

# Ensemble by averaging probabilities
ensemble_probs = np.mean(model_probs, axis=0)
ensemble_preds = np.argmax(ensemble_probs, axis=1)

# Convert true labels to numpy array
true_labels = test_labels.cpu().numpy() if torch.is_tensor(test_labels) else np.array(test_labels)

# Binarize true labels for multiclass ROC-AUC
n_classes = 4
true_labels_bin = label_binarize(true_labels, classes=np.arange(n_classes))

# Calculate ROC-AUC score (macro average)
roc_auc = roc_auc_score(true_labels_bin, ensemble_probs, average='macro', multi_class='ovr')
print(f'Ensemble ROC-AUC Score (macro): {roc_auc:.4f}')

# Plot ROC curve for each class
plt.figure(figsize=(8, 6))
fpr = dict()
tpr = dict()
roc_auc_class = dict()
class_names = ['not bully', 'religious', 'others', 'sexual']

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(true_labels_bin[:, i], ensemble_probs[:, i])
    roc_auc_class[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'{class_names[i]} (AUC = {roc_auc_class[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random guess')
plt.title('ROC Curve per Class (Ensemble)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.savefig("ensemble_english_roc_curve.png", dpi=300)
plt.show()

# Calculate MCC and Cohen's Kappa
mcc = matthews_corrcoef(true_labels, ensemble_preds)
kappa = cohen_kappa_score(true_labels, ensemble_preds)

print(f'Matthews Correlation Coefficient (MCC): {mcc:.4f}')
print(f"Cohen's Kappa Score: {kappa:.4f}")
```
